## Praktikum

### a) Import modul 
Tulis kode ke dalam *cell* di bawah ini:


```python
import numpy as np
```

### b) Fungsi Aktivasi

Tulis kode ke dalam *cell* di bawah ini:

```python
def aktivasi(x):
  if x < 0:
    return -1
  else:
    return 1
```

### c) Fungsi *Training* Madaline

Tulis kode ke dalam *cell* di bawah ini:

```python
def train(train_data, train_target, alpha=0.1, max_epoch=10):
  w = np.random.random((2, 2))
  v = np.array([0.5, 0.5])
  b = np.random.random(2)
  b = np.append(b, 0.5)
  epoch = 0
  v_aktivasi = np.vectorize(aktivasi)
  weight_updated = True
  
  while weight_updated and epoch < max_epoch:
      weight_updated = False
      for data, target in zip(train_data, train_target):
          z_in = np.dot(data, w)
          z_in = z_in + b[:-1]
          z = v_aktivasi(z_in)
          y_in = np.dot(z, v) + b[-1]
          y = v_aktivasi(y_in)

          if y != target:
              weight_updated = True
              if target == 1:
                  index = np.argmin(np.abs(z_in))
                  b[index] = b[index] + alpha * (1 - z_in[index])
                  w[:, index] = w[:, index] + alpha * (1 - z_in[index]) * data
              elif target == -1:
                  index = np.where(z_in > 0)[0]
                  if len(index) == 1:
                      index = index[0]
                      b[index] = b[index] + alpha * (-1 - z_in[index])
                      w[:, index] = w[:, index] + alpha * (-1 - z_in[index]) * data
      epoch += 1
  
  return (w, v, b)
```

### d) Fungsi *Testing* Madaline

Tulis kode ke dalam *cell* di bawah ini:

```python
def test(w,v,b,test_data):
  v_aktivasi = np.vectorize(aktivasi)
  z_in = np.dot(test_data, w)
  z_in = z_in + b[:-1]
  z = v_aktivasi(z_in)
  y_in = np.dot(z, v) + b[-1]
  y = v_aktivasi(y_in)
  return y
```

### e) Fungsi Hitung Akurasi

Tulis kode ke dalam *cell* di bawah ini:

```python
def calc_accuracy(a, b):
  s = [1 if a[i] == b[i] else 0 for i in range(len(a))]
  return sum(s) / len(a)
```

### f) Logika AND

Tulis kode ke dalam *cell* di bawah ini:

```python
data = np.array([[1,1],[1,-1],[-1,1],[-1,-1]])
target = np.array([1,-1,-1,-1])
(w,v,b) = train(data,target,alpha=1,max_epoch=10)
output = test(w,v,b,data)
accuracy = calc_accuracy(output, target)
print('Output:', output)
print('Target:', target)
print('Accuracy:', accuracy)
```

### g) Logika OR

Tulis kode ke dalam *cell* di bawah ini:

```python
data = np.array([[1,1],[1,-1],[-1,1],[-1,-1]])
target = np.array([1,1,1,-1])
(w,v,b) = train(data,target,alpha=0.2,max_epoch=10)
output = test(w,v,b,data)
accuracy = calc_accuracy(output, target)
print('Output:', output)
print('Target:', target)
print('Accuracy:', accuracy)
```

### h) Logika XOR

Tulis kode ke dalam *cell* di bawah ini:

```python
data = np.array([[1,1],[1,-1],[-1,1],[-1,-1]])
target = np.array([-1,1,1,-1])
(w,v,b) = train(data,target,alpha=0.7,max_epoch=10)
output = test(w,v,b,data)
accuracy = calc_accuracy(output, target)
print('Output:', output)
print('Target:', target)
print('Accuracy:', accuracy)
```

### h) Logika OR Biner

Tulis kode ke dalam *cell* di bawah ini:

```python
import numpy as np

# Fungsi aktivasi untuk output biner
def aktivasi(x):
    return 1 if x >= 0 else 0

# Fungsi untuk melatih jaringan
def train(train_data, train_target, alpha=0.1, max_epoch=10):
    w = np.random.random((2, 2))  # Bobot untuk neuron dalam lapisan tersembunyi
    v = np.array([0.5, 0.5])       # Bobot untuk neuron output
    b = np.random.random(2)         # Bias untuk neuron dalam lapisan tersembunyi
    b = np.append(b, 0.5)           # Bias untuk neuron output
    epoch = 0
    v_aktivasi = np.vectorize(aktivasi)
    weight_updated = True

    while weight_updated and epoch < max_epoch:
        weight_updated = False
        for data, target in zip(train_data, train_target):
            z_in = np.dot(data, w) + b[:-1]  # Input ke neuron dalam lapisan tersembunyi
            z = v_aktivasi(z_in)               # Output dari neuron dalam lapisan tersembunyi
            y_in = np.dot(z, v) + b[-1]       # Input ke neuron output
            y = aktivasi(y_in)                 # Output dari neuron output

            if y != target:
                weight_updated = True
                if target == 1:
                    index = np.argmin(np.abs(z_in))
                    b[index] += alpha * (1 - z_in[index])
                    w[:, index] += alpha * (1 - z_in[index]) * data
                elif  target == 0:  # target == 0
                    index = np.where(z_in > 0)[0]
                    if len(index) == 1:
                        index = index[0]
                        b[index] += alpha * (-1 - z_in[index])
                        w[:, index] += alpha * (-1 - z_in[index]) * data
        epoch += 1

    return (w, v, b)

# Fungsi untuk menguji jaringan
def test(w, v, b, test_data):
    v_aktivasi = np.vectorize(aktivasi)
    z_in = np.dot(test_data, w) + b[:-1]  # Input ke neuron dalam lapisan tersembunyi
    z = v_aktivasi(z_in)                    # Output dari neuron dalam lapisan tersembunyi
    y_in = np.dot(z, v) + b[-1]            # Input ke neuron output
    y = v_aktivasi(y_in)                    # Output dari neuron output
    return y

# Fungsi untuk menghitung akurasi
def calc_accuracy(output, target):
    return np.mean(output == target)

# Data dan target untuk logika OR
data = np.array([[1, 1], [1, 0], [0, 1], [0, 0]])
target = np.array([1, 1, 1, 0])  # Target biner untuk logika OR

# Melatih jaringan
(w, v, b) = train(data, target, alpha=0.75, max_epoch=10)

# Menguji jaringan
output = test(w, v, b, data)
accuracy = calc_accuracy(output, target)

# Menampilkan hasil
print('Output:', output)
print('Target:', target)
print('Accuracy:', accuracy)
```
