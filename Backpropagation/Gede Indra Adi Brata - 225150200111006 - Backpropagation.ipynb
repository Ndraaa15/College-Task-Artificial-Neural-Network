Pembuatan Fungsi

```python
def binary_encoder(column):
    unique_values = column.unique()
    binary_map = {val: format(i, f'0{len(bin(len(unique_values)-1)[2:])}b') for i, val in enumerate(unique_values)}
    encoded_column = column.map(binary_map)
    return encoded_column, binary_map

def binary_decoder(encoded_column, binary_map):
    reverse_map = {v: k for k, v in binary_map.items()}
    decoded_column = encoded_column.map(reverse_map)
    return decoded_column


data['Marital Status'], marital_status_map = binary_encoder(data['Marital Status'])
print(marital_status_map)

data['Gender'], gender_map = binary_encoder(data['Gender'])
print(gender_map)

data['Home Owner'], home_owner_map = binary_encoder(data['Home Owner'])
print(home_owner_map)

data['Purchased Bike'], purchased_bike_map = binary_encoder(data['Purchased Bike'])
print(purchased_bike_map)
```

Kode di atas terdiri dari dua fungsi, binary_encoder dan binary_decoder, yang digunakan untuk mengonversi data kategorikal menjadi representasi biner dan mengembalikannya ke nilai asli. Fungsi binary_encoder menerima sebuah kolom data sebagai input, lalu menemukan nilai-nilai unik dalam kolom tersebut. Selanjutnya, fungsi ini membuat peta (dictionary) yang mengaitkan setiap nilai unik dengan kode biner yang sesuai berdasarkan urutan indeksnya. Kode biner disusun dengan panjang yang cukup untuk merepresentasikan semua nilai unik dalam bentuk biner. Kolom data asli kemudian dikonversi ke kode biner menggunakan peta ini dan dikembalikan bersama dengan peta biner. Fungsi binary_decoder menerima kolom yang telah dikodekan dan peta biner dari binary_encoder untuk mengembalikan data ke bentuk awal. Ini dilakukan dengan membuat peta terbalik, lalu memetakan nilai biner ke nilai asli berdasarkan peta tersebut.

```python
def ordinal_encoder(column, hierarchy=None):
    unique_values = hierarchy if hierarchy else sorted(column.unique())
    ordinal_map = {val: i for i, val in enumerate(unique_values)}
    encoded_column = column.map(ordinal_map)
    return encoded_column, ordinal_map

def ordinal_decoder(encoded_column, ordinal_map):
    reverse_map = {v: k for k, v in ordinal_map.items()}
    decoded_column = encoded_column.map(reverse_map)
    return decoded_column


education_hierarchy = ['Partial High School', 'High School', 'Partial College', 'Bachelors', 'Graduate Degree']
commute_distance_hierarchy = ['0-1 Miles', '1-2 Miles', '2-5 Miles', '5-10 Miles', '10+ Miles']

data['Education'], education_map = ordinal_encoder(data['Education'], hierarchy=education_hierarchy)
print(education_map)

data['Commute Distance'], commute_distance_map = ordinal_encoder(data['Commute Distance'], hierarchy=commute_distance_hierarchy)
print(commute_distance_map)
```

Kode di atas terdiri dari dua fungsi, ordinal_encoder dan ordinal_decoder, yang digunakan untuk mengonversi data kategorikal secara ordinal sesuai dengan hierarki yang diinginkan, dan mengembalikannya ke bentuk asli. Fungsi ordinal_encoder menerima kolom data dan opsional hierarchy sebagai input. Jika hierarchy diberikan, fungsi akan menggunakan urutan nilai dari hierarki tersebut; jika tidak, nilai unik dalam kolom akan diurutkan secara alfabetis. Fungsi kemudian membuat peta ordinal (ordinal_map) yang mengasosiasikan setiap nilai unik dengan indeks berdasarkan urutan yang telah ditentukan. Kolom asli dikonversi menjadi representasi ordinal berdasarkan peta ini, dan fungsi mengembalikan kolom hasil encoding beserta peta ordinalnya. Fungsi ordinal_decoder digunakan untuk mengembalikan kolom hasil encoding ke bentuk awal dengan membalik peta ordinal, lalu memetakan nilai ordinal kembali ke nilai asli sesuai peta terbalik ini.

```python
def one_hot_encoder(column):
    unique_values = column.unique()
    one_hot_map = {val: [1 if i == j else 0 for i in range(len(unique_values))] for j, val in enumerate(unique_values)}
    encoded_columns = pd.DataFrame([one_hot_map[val] for val in column], columns=unique_values)
    return encoded_columns, one_hot_map

def one_hot_decoder(encoded_columns, one_hot_map):
    reverse_map = {tuple(v): k for k, v in one_hot_map.items()}
    decoded_column = encoded_columns.apply(lambda row: reverse_map[tuple(row)], axis=1)
    return decoded_column

occupations, occupation_map = one_hot_encoder(data['Occupation'])
print(occupation_map)

regions, region_map = one_hot_encoder(data['Region'])
print(region_map)

data = pd.concat([data, occupations], axis=1)
data = pd.concat([data, regions], axis=1)
data = data.drop(columns=['Occupation', 'Region'])
```

Kode di atas terdiri dari dua fungsi, one_hot_encoder dan one_hot_decoder, yang digunakan untuk encoding dan decoding data kategorikal menggunakan one-hot encoding. Fungsi one_hot_encoder mengambil kolom data sebagai input dan menentukan nilai unik dalam kolom tersebut. Untuk setiap nilai unik, fungsi membuat peta (one_hot_map) yang mengasosiasikan setiap nilai dengan vektor biner, di mana elemen pada indeks yang sesuai dengan nilai tersebut diatur menjadi 1, sementara elemen lainnya diatur menjadi 0. Kolom asli kemudian dikonversi ke dalam DataFrame dengan format one-hot encoding, yang mengandung satu kolom untuk setiap nilai unik, dan fungsi mengembalikan DataFrame hasil encoding beserta peta one-hot-nya. Fungsi one_hot_decoder mengembalikan DataFrame yang telah di-encode ke bentuk awalnya dengan menggunakan peta terbalik (reverse_map), yang mencocokkan setiap vektor biner ke nilai aslinya, lalu menerapkan peta tersebut pada setiap baris DataFrame.
